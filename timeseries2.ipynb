```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("PANW.csv")
df.head()
```

```python
df1 = df.drop(columns=(['Open', 'High', 'Low', 'Close', 'Volume']))
df1.set_index('Date', inplace=True)
df1.head()
```

```python
plt.figure(figsize=(50, 15))
plt.plot(df1.index, df1, color='blue', linestyle='-')
plt.title('Price', fontsize=20)
plt.xlabel('Date', fontsize=16)
plt.ylabel('Logarithmic Returns', fontsize=16)

tick_ratio = 0.025  # Set the desired ratio of x ticks
n_ticks = int(len(df1.index) * tick_ratio)
tick_positions = range(0, len(df1.index), n_ticks)
tick_labels = df1.index[tick_positions]
plt.yticks(fontsize=14)
plt.xticks(tick_positions, tick_labels, rotation=45, fontsize=14)

plt.show()
```

```python
log_returns = df1['Adj Close'].apply(lambda x: np.log(x)).diff()

# Drop the first row (NaN due to no previous value)
log_returns = pd.DataFrame(log_returns.dropna())
log_returns.head()

plt.figure(figsize=(50, 15))
plt.plot(log_returns.index, log_returns, color='blue', linestyle='-')
plt.title('Logarithmic Returns', fontsize=20)
plt.xlabel('Date', fontsize=16)
plt.ylabel('Logarithmic Returns', fontsize=16)

tick_ratio = 0.025  # Set the desired ratio of x ticks
n_ticks = int(len(log_returns.index) * tick_ratio)
tick_positions = range(0, len(log_returns.index), n_ticks)
tick_labels = log_returns.index[tick_positions]
plt.yticks(fontsize=14)
plt.xticks(tick_positions, tick_labels, rotation=45, fontsize=14)

plt.show()
```
